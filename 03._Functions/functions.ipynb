# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def main(documents,directories):
    while True:
        comm=input('Введите комманду: ')
        comm=comm.lower()
        if comm=='p':
            result = check_dock(userinp('docnum'))
            if result[0] != None: print(f"Владелец документа: {documents[result[0]]['name']}")
        elif comm=='s': 
            result = check_dock(userinp('docnum'))
            if result[1] != None: print(f"Документ хранится на полке: {result[1]}")
        elif comm=='l': 
            p3_getall()
        elif comm=='ads': 
            p4_adddir()
        elif comm=='ds': 
            p5_deldir()
        elif comm=='ad': 
            p21_adddoc()
        elif comm=='d': 
            p22_deldoc()
        elif comm=='m': 
            p23_movedoc()
        elif comm=='q':
            break
        else:
            print('Неизвестная комманда')
main(documents,directories)
````

````python
def userinp (inptype):
    """
    Запрашивает ввод от пользователя с соответствующим сообщением
    Если нужен ввод числа, делает это с проверкой корректности ввода
    Возвращает результат ввода
    """
    if inptype=='docnum':
        mes=('Введите номер документа: ')
    elif inptype=='dirnum':
        mes=('Введите номер полки: ')
    elif inptype=='doctype':
        mes=('Введите тип документа:')
    elif inptype=='docname':
        mes=('Введите владельца документа:')
    
    inpstr=input(mes)
    if inptype=='dirnum':
        while not inpstr.isnumeric():
            inpstr=input('Некорректный ввод числа, повторите ввод: ')
        return int(inpstr)    
    else:   
        return inpstr 
#userinp('doctype')  

def check_dock (num,silent=False):
    """
    Проверяет наличие документа с номером num в перечне документов и в перечне полок
    Возвращает список [порядковый номер документа в списке, номер полки]
    Если документ не найден, возвращает [None,None]
    При silent=False (по умолчанию False) выводит сообщение если документ не найден
    Если документ не найден только в одном из перечней, выводит ошибку
    """
    result=[None,None]
    for dir in directories:
        if num in directories[dir]:
            result[1]=int(dir)
    for doc in enumerate(documents):
        if doc[1]['number']==num:
            result[0]=doc[0]
    if None in result:
        if result==[None,None]:
            if not silent: print('Документ не найден в базе')
        else:
            print('Критическая ошибка: база повреждена')
    return result
#print(check_dock('2207 876234'))
            
    
def dir_exist (num,silent=False):
    result = str(num) in directories.keys()
    if not result and not silent:
        print('Такой полки не существует. '+dirlist()) 
    return result
     
def dirlist ():
    return('Текущий перечень полок: '+', '.join(directories.keys()))
````

````python
def p3_getall ():
    print('Текущий список документов:')
    for doc in documents:
         print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {check_dock(doc['number'])[1]}") 
#p3_getall()

def p4_adddir ():
    dirnum = userinp ('dirnum')
    if dir_exist(dirnum,silent=True):
        print('Такая полка уже существует. '+dirlist()) 
    else:
        directories[str(dirnum)]=[]
        print('Полка добавлена. '+dirlist()) 
#p4_adddir()

def p5_deldir ():
    dirnum = userinp ('dirnum')
    if dir_exist(dirnum):
        if len(directories[str(dirnum)])>0:
            print('На полке есть документы, удалите их перед удалением полки. '+dirlist()) 
        else:
            directories.pop(str(dirnum))
            print('Полка удалена. '+dirlist()) 
#p5_deldir()

def p21_adddoc ():
    newdoc={};
    newdoc['number'] = userinp('docnum')
    if not (check_dock(newdoc['number'],silent=True) == [None,None]):
        print('Документ с таким номером уже есть в базе')
    else:
        newdoc['type'] = userinp('doctype')
        newdoc['name'] = userinp('docname')
        dirnum = userinp ('dirnum')
        if not dir_exist(dirnum,silent=True):
            print('Такой полки не существует. Добавьте полку командой as. ')      
        else:
            documents.append(newdoc)
            directories[str(dirnum)].append(newdoc['number'])
            print('Документ добавлен. Текущий список документов:')
        p3_getall()       
#p21_adddoc()

def p22_deldoc ():
    docnum=userinp('docnum')
    docind=check_dock(docnum)
    if not (None in docind):
        documents.pop(docind[0])
        directories[str(docind[1])].remove(docnum)
        print('Документ удален.')
    p3_getall()  
#p22_deldoc()

def p23_movedoc ():
    docnum=userinp('docnum')
    olddir=check_dock(docnum)[1]
    if olddir!=None:
        newdir = userinp ('dirnum')
        if dir_exist(newdir) :
            olddir=str(olddir)
            directories[str(newdir)].append( directories[olddir].pop(directories[olddir].index(docnum)) )
            print('Документ перемещен.')
            p3_getall()
    else:        
        p3_getall()   
#p23_movedoc()
````
